```python
df = pd.read_csv('Public.csv', encoding = 'EUC-KR')
```

```python
df.info()
```

```python
df.head()
```

```python
#준공일자가 int니까 준공일자로 정렬해 보자
df.sort_values( '준공일자', ascending= True)
```

```python
# 준공일자를 datetime으로 바꿔보자
from datetime import datetime
df['준공일자'] = pd.to_datetime(df['준공일자'], format = '%Y%m%d')
```

```python
df['Year'] = df['준공일자']
```

```python
#이런 식으로 항상 기준을 정하는 게 중요하다
#아까 년도가 1900년부터 시작이었는데 제대로 정렬이 안 됨.
df.set_index('준공일자', inplace = True)
df.head()
```

```python
df.sort_values('준공일자', ascending = True)
```

```python
# 현재 인덱스를 보면 준공일자로 되어 있는데, 아까 datetime으로 변경해줄 떄 봤겠지만
#%Y%m%d 순으로 적어줬음. 3개가 합쳐진 거라는 소리.
df['준공년도'] = df.index.year; df.head()
```

```python
# 각 년도별 준공건수를 찾아보자.
df_year = df['준공년도'].value_counts();df_year
```

```python
# 준공 수가 많은 년도로 정렬이 되어 있는데 년도별로 정렬을 해보자.
df_year.sort_index().head()
```

```python
# 시설물도 시, 도별로 구분을 해보자.
sample = df['시설물소재지'].values[0]; sample
```

```python
#위 문자열을 샘플로 한 번 부산광역시까지 추출을 해보자
sample.find(' ')
```

```python
# 첫 띄어쓰기가 중요하다. 여기까지 딱 나온다.
# 그럼 df['시설물소재지'].values[0][:sample.find(' ')]이렇게 하면 부산광역시가 나오는지 확인해 보자
df['시설물소재지'].values[0][:sample.find(' ')]
# 리스트 컴프리헨션을 쓰면 너무 더러울 것 같으니 그냥 for문 돌리자.
```

```python
# 일단 빈 공간을 만들어줘야 할 것 같다. 그냥 포문을 돌리니 에러가 난다.
df['state'] = str()
```

```python
# 시, 군만 추출하는 for문 생성
# 주의해야 할 것들은 index는 index, value는 value처럼 내가 건들일 값의 소재지들을
# 명확하게 구분을 해줘야 한다.
#df['state'].[i] = df['시설물소재지'].values[i][:df['시설물소재지'].values[i].find(' ')]
# 이런 건 안 된다.
# datetime은 loc이 안 돼서 iloc으로 인덱스 번호만 빼와야 한다.
for i in range(len(df)):
    df['state'].values[i] = df['시설물소재지'].values[i][:df['시설물소재지'].values[i].find(' ')]
#    df.iloc[:, 5].values[i] = df['시설물소재지'].values[i][:df['시설물소재지'].values[i].find(' ')]
df.head()
```

```python
# 오래된 기록들이니 혹시 서울시, 서울특별시 이렇게 나뉘어지지는 않았는지 확인해 보자.
df['state'].unique()
```

```python
# 이제 시, 도별 준공건수를 세어보자!
df_state = df['state'].value_counts(); df_state
```

```python
plt.figure(figsize=(20,20))
plt.subplot(211)
plt.title('노후공공시설물 설립년도별 현황')
df_year.plot(kind='bar')
plt.subplot(223)
plt.title("시도별 노후공공시설 현황")
df_state.plot(kind = 'Pie')
plt.subplot(224)
plt.title("시도별 노후공공시설 순위 TOP10")
df_state.sort_values(ascending=True).head(10).plot(kind='barh', color='y')
plt.show()
```
