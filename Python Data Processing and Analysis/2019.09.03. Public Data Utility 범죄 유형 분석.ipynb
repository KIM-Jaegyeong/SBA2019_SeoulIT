```python
# 범죄대분류 별 건수
df_crime = pd.Series(df['계'].values, index=df['범죄대분류'].values)
df_crime
```

```python
def get_crime_cnt(d_frame):
    crime_cnt = {}
    for idx in range(len(d_frame.index)):
        # print(idx, d_frame.index[idx], d_frame[idx], end="  \t =>")
        # print(idx, crime_cnt)

        crime = d_frame.index[idx]
        count  = d_frame[idx]

        if crime_cnt.get(crime):
            crime_cnt[crime] += count
        else:
            crime_cnt[crime] = count
    
    return crime_cnt
    
crime_dict = get_crime_cnt(df_crime)
crime_dict
```

```python
# reduce로 대분류 갯수 알아내기

from functools import reduce
# a가 {}이고 b가 key, values임
def draw_histogram(a, b):
     if not a.get(b[0]):
         a[b[0]] = b[1]
     else:
         a[b[0]] += b[1]
     return a
     print(a, b)
        
reduce(draw_histogram, df_crime.items(), {})
```
