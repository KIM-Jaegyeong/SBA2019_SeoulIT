# Merge

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.randn(10, 4))
df
```

```python
# break it into pieces
pieces = [df[:3], df[3:7], df[7:]]
pd.concat(pieces) # 디폴트 0이라  axis 0으로 붙음
```

```python
pd.concat(pieces, axis = 1)
```

# Join

```python
left = pd.DataFrame({'key': ['foo', 'bar'], 'lval': [1, 2]})
right = pd.DataFrame({'key': ['foo', 'bar'], 'rval': [4, 5]})
```

```python
left
```

```python
right
```

```python
pd.concat([left, right])
```

```python
# key를 기준으로 right부터 오름차순 기준 작은 인덱스에 해당하는 key값을 넣고 그 키값에 해당하는
# rval과 lval을 순서대로 가져옴
pd.merge(right, left, on = 'key')
```

# Grouping

```python
df = pd.DataFrame({'A': ['foo', 'bar', 'foo', 'bar',
   ....:                          'foo', 'bar', 'foo', 'foo'],
   ....:                    'B': ['one', 'one', 'two', 'three',
   ....:                          'two', 'two', 'one', 'three'],
   ....:                    'C': np.random.randn(8),
   ....:                    'D': np.random.randn(8)})
df
```

```python
df.groupby('A').sum() # A컬럼의 종류는 bar, foo인데 데이터값을 가진 c와 d의 합을 알려줌
```

```python
df.groupby(['A', 'B']).sum() # bar 중에 one의 합
#                                      three의 합...
```

```python
df.groupby(['B', 'A']).sum() # bar 중에 one의 합
#                                      three의 합...
```
