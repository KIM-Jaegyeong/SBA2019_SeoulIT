# 넘파이 사용

```python
# sin-1.py

import math
import matplotlib.pyplot as plt

T = range(100)
X = [(2 * math.pi * t) / len(T) for t in T] #  0부터 99까지로 나눠주고 곱해줌
Y = [math.sin(value) for value in X] # X의 값을 sin에 넣음

plt.plot(X, Y)
plt.show()
```

```python
# sin-2.py

import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(0, 2*np.pi, 100) # 0부터 2*np.pi를 100개로 잘라낸 거
Y = np.sin(X)

plt.plot(X, Y)
plt.show()
```

# 넘파이 패키지

```python
# 이항식 : x^2 - 2x + 1

import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-3, 2, 200) # -3부터 -2까지 200개로 잘라냄
Y = X**2 - 2*X + 1.

plt.plot(X, Y)
plt.show()
```

# 다중 곡선 그리기

```python
# [0, 2*pi] 간격의 sin(x)과 cos(x)을 둘 다 보여주기

import numpy as np
import matplotlib.pyplot as plt

X  = np.linspace(0, 2*np.pi, 100) #100등분이 기본값
Ya = np.sin(X)
Yb = np.cos(X)

plt.plot(X, Ya, 'r') # sin의 그래프 
plt.plot(X, Yb) # 
plt.show()
```

```python
# 지연 랜더링, deferred rendering 메커니즘은 맷플롯립의 핵심

import numpy as np
import matplotlib.pyplot as plt

def plot_slope(X, Y):
    Xs = X[1:] - X[:-1]
    Ys = Y[1:] - Y[:-1]
    plt.plot(X[1:], Ys/Xs)
    
X = np.linspace(-3, 3, 100)
Y = np.exp(-X**2)

plt.plot(X, Y)
plot_slope(X, Y)
plt.show()
```

```python
# 여러 수량을 비교
import numpy as np
import matplotlib.pyplot as plt

data = [[5., 25., 50., 20.],
        [4., 23., 51., 17.],
        [6., 22., 52., 19.]]

X = np.arange(4)
plt.bar(X + 0.00, data[0], color='b', width=0.25)
plt.bar(X + 0.25, data[1], color='g', width=0.25)
plt.bar(X + 0.50, data[2], color='r', width=0.25)

plt.show()
```

```python
# 반복자 enumerate 는 현재 행과 해당 첨자를 둘 다 반환
import numpy as np
import matplotlib.pyplot as plt

data = [[5., 25., 50., 20.],
        [4., 23., 51., 17.],
        [6., 22., 52., 19.]]

color_list = ['b', 'g', 'r']
gap = .8 / len(data)
for i, row in enumerate(data): # 0, [5., 25., 50., 20.] 이런 식으로 들어감
    X = np.arange(len(row))
    plt.bar(X + i*gap, row, 
           width = gap, 
           color = color_list[i % len(color_list)])
plt.show()
```

# 분할 막대 차트 그리기

```python
import matplotlib.pyplot as plt

A = [5., 30., 45., 22.]
B = [5., 25., 50., 20.]

X = range(4)

plt.bar(X, A, color='b')
plt.bar(X, B, color='r', bottom=A)# 그릴 그래프가 그 전의 그래프 위에 쌓이게 하고 싶다면
plt.show()
```

```python
# 값을 두 집합 이상으로 분할
import numpy as np
import matplotlib.pyplot as plt

A = np.array([5., 30., 45., 22.])
B = np.array([5., 25., 50., 20.])
C = np.array([1.,  2.,  1.,  1.])
X = range(4)

plt.bar(X, A, color='b')
plt.bar(X, B, color='g', bottom=A)
plt.bar(X, C, color='r', bottom=A+B)
plt.show()
```

```python
# 값을 두 집합 이상으로 분할
import numpy as np
import matplotlib.pyplot as plt

data = np.array([[5., 30., 45., 22.],
                 [5., 25., 50., 20.],
                 [1.,  2.,  1.,  1.]        
    ])

color_list = ['b', 'g', 'r']

X = np.arange(data.shape[1])

for i in range(data.shape[0]): # 이러면 shape가 4*3이니까 0, 1, 2가 들어감
    plt.bar(X, data[i],# X는 0, 1, 2고 data[i]는 순서대로 [5., 30., 45., 22.]...
           bottom = np.sum(data[:i], axis=0),# axis가 0이라 컬럼별로 더해짐
           color = color_list[i % len(color_list)])

plt.show()
```

# 삼각화 그리기

```python
# 공간 위치를 처리할 때 삼각화 생긴다. 점과 이웃 관계 사이의 거리를 보여줌
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.tri as tri

#data = np.random.rand(100, 2)
data = np.array([[1, 3], [5, 4], [4, 2]])
triangles = tri.Triangulation(data[:,0], data[:,1])

plt.triplot(triangles)
plt.show()
```
